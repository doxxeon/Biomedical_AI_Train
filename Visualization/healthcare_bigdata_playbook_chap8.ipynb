노약자 위치 데이터로 plot하기 !

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('OTg6QzA_activities.csv')
print(df.head())
print(df.shape)
```

```python
X = df.X.to_numpy()
print(X[0])
Y = df.Y.to_numpy()
print(Y[0])

plt.scatter(X,Y)
plt.show()
```

```python
plt.rc('figure', figsize=(12, 10))
```

```python
# 위치 정보를 2D로 찍기 위해!

x1 = X[:16570]
x2 = X[16570:16570*2]
x3 = X[16570*2:16570*3]
x4 = X[16570*3:]

y1 = Y[:16570]
y2 = Y[16570:16570*2]
y3 = Y[16570*2:16570*3]
y4 = Y[16570*3:]
```

```python
plt.figure(facecolor='white')

# 1등분
plt.subplot(2,2,1)
plt.hist2d(x1, y1, bins=20, cmap='viridis')
plt.title('1-Q')
plt.colorbar()

# 2등분
plt.subplot(2,2,2)
plt.hist2d(x2, y2, bins=20, cmap='plasma')
plt.title('2-Q')
plt.colorbar()

# 3등분
plt.subplot(2,2,3)
plt.hist2d(x3, y3, bins=20, cmap='magma')
plt.title('3-Q')
plt.colorbar()

# 4등분
plt.subplot(2,2,4)
plt.hist2d(x4, y4, bins=20, cmap='inferno')
plt.title('4-Q')
plt.colorbar()

plt.show()
```

나중에는 침대 근처가 아닌 다른 곳에 많이 찍혔기 때문에 회복이 되었다고도 분석할 수 있음

```python
plt.figure(facecolor='white')


# 1등분
plt.subplot(2,2,1)
#plt.hist2d(x1, y1, bins=20, cmap='viridis')
plt.scatter(x1, y1, color='black', s=0.1)
plt.title('1-Q')
plt.colorbar()

# 2등분
plt.subplot(2,2,2)
#plt.hist2d(x2, y2, bins=20, cmap='plasma')
plt.scatter(x2, y2, color='red', s=0.1)
plt.title('2-Q')
plt.colorbar()

# 3등분
plt.subplot(2,2,3)
#plt.hist2d(x3, y3, bins=20, cmap='magma')
plt.scatter(x3, y3, color='blue', s=0.1)
plt.title('3-Q')
plt.colorbar()

# 4등분
plt.subplot(2,2,4)
#plt.hist2d(x4, y4, bins=20, cmap='inferno')
plt.scatter(x4, y4, color='pink', s=0.1)
plt.title('4-Q')
plt.colorbar()


plt.show()
```

```python
plt.scatter(x1, y1, color='black', s=1)
plt.scatter(x2, y2, color='red', s=1)
plt.scatter(x3, y3, color='blue', s=1)
plt.scatter(x4, y4, color='pink', s=1)
```
